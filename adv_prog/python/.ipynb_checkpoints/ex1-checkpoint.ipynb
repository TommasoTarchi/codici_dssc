exercise 1

```python
def recaman(N):
    rec = [0]
    while len(rec) < N:
        n = len(rec)
        a = rec[n-1]-n
        if a > 0 and not a in rec:
            rec.append(a)
        else:
            rec.append(rec[n-1]+n)
    return rec

recaman(18)
```

exercise 3

```python
mylist = [1,-99,89,0,234,77,-748,11]
final = []

while len(final) != len(mylist):
    p = -1000
    for i in range(len(mylist)):
        if (not mylist[i] in final) and mylist[i] > p:
            p = mylist[i]
    final.append(p)

print(final)
```

exercise 5

```python
def factors(N):
    factors = [(x, y) for x in range(N+1) for y in range(N+1) if x*y == N if x<=y]
    return factors

print(factors(12))
```

exercise 6

```python
def first():
    num = []
    for n in range(3, 1000):
        if n%3 == 0:
            num.append(n)
    return num

def second():
    return [n for n in range(3, 1000) if n%3 == 0]

def third():
    return list(range(3, 1000))[::3]
    
def fourth():
    return list(range(3, 1000, 3))


print(first())
print(second())
print(third())
print(fourth())
```

exercise 7 (not distinguishing lower and upper case)

```python
def pal(s):
    s = s.lower()
    return s[0:len(s)//2] == s[len(s)-1:len(s)-len(s)//2-1:-1]

print(pal("Anna"))
```

exercise 8

```python
def max_occ(s):
    s = s.lower()
    letters = set(s)
    occ = ['', 0]
    for lett in letters:
        if s.count(lett) > occ[1]:
            occ = [lett, s.count(lett)]
    return occ

print(max_occ('sSsjasn'))
```

exercise 9

```python
def erat(N):
    primes = []
    n = 2
    while len(primes) != N:
        control = 1
        for p in primes:
            if n%p == 0:
                control = 0
        if control:
            primes.append(n)
        n = n+1
    return primes

print(erat(8))
```
