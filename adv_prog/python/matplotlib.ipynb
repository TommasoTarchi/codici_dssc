```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn
```

```python
seaborn.set_theme()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.title('Plot of some sinusoidal stuff', fontsize = 'xx-large', fontstyle='italic', fontweight = 'light')
plt.ylabel('This is the $y$ axis', fontsize = 'xx-large', fontstyle='oblique', fontweight = 'heavy')
plt.xlabel('This is the $x$ axis', fontsize = 'xx-large', fontstyle='normal', fontweight = 'medium')

plt.plot(x, y1)
plt.plot(x, y2)
```

```python
plt.plot(x, y1, linestyle='-.', linewidth=3, color=(139/255, 0, 139/255), marker=('s'))
```

```python
x = np.linspace(0,5,20)
y1 = np.sin(x)
y2 = np.sin(x + 0.2)
y3 = np.sin(x + 0.4)
y4 = np.sin(x + 0.6)
y5 = np.sin(x + 0.8)
y6 = np.sin(x + 1.0)
y7 = np.sin(x + 1.2)

# now we decide the actual figure size in inches
fig = plt.figure(figsize=(12,6))

# create subplots now we have created a 2x2 grid and a 4x4 grid and fitted them together
plt.subplot(221)
plt.plot(x,y1,color='r')
plt.subplot(223)
plt.plot(x,y2,color='g')
plt.subplot(224)
plt.plot(x,y7,color='k')

plt.subplot(443)
plt.plot(x,y3,color='b')
plt.subplot(444)
plt.plot(x,y4,color='y')
plt.subplot(447)
plt.plot(x,y5,color='c')
plt.subplot(448)
plt.plot(x,y6,color='m')
```

posso definire le variabili subplot, le quali hanno una serie di attributi (lo stesso .plot()  meglio vederlo come un metodo per assegnare il 'contenuto' al plot piuttosto che un comando per plottare graficamente)

```python
asc = np.linspace(0, 10, 100)

g = plt.subplot(121)
h = plt.subplot(122)
g.plot(asc, np.sin(asc))
h.plot(asc, np.cos(asc))
g.set_aspect('equal')
```
