```python
def call(f, vect):
    if vect.size != 0:
        return f(vect)
    else:
        pass
```

```python
def func(v):
    print(sum(v))

call(func, np.array([]))
call(func, np.array([1, 3, 6]))
```

#### ex. 3

```python
def add_time(f):

    # avrei assegnare f(...) ad una variabile e poi ritornare quella variabile (per non dover
    # chiamare due volte la funzione), ma in quel modo falserei il tempo misurato perch√©
    # starei sommando il tempo di assegnazione della variabile
    def wrapper(*args, **kwargs):
        start=time.time()
        f(*args, *kwargs)
        stop=time.time()
        print("function time is: " + str(stop-start))
        return f(*args, **kwargs)

    return wrapper
```

```python
@add_time
def func(x):
    return (x+5)**10

print(func(5))

@add_time
def func_sleep():
    time.sleep(5)

func_sleep()
```

```python
def plot_function(f):

    interval = np.linspace(0, 10, 1000)

    plt.plot(interval, f(interval))
```

```python
plot_function(np.sin)
```
